# Matplotlib example workflow

```python
# 0. import matplotlib and inline
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [20,10,40,30]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) #(width, height)

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. save and show
fig.savefig('sample-fig.png')
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.rand(10), np.random.rand(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(100));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.rand(10), np.random.rand(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(100));
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease
```

### Advanced plots use OO method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# PyPlot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO Method mixed with pyPlot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

ax.set_xlim([50,80])
```

```python
# Object Oriented Method from scratch
fog, ax = plt.subplots(figsize=(10,6))

# Plot data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# Customize plot
ax.set(title='Heart Disease Gang',
      xlabel='Age',
      ylabel='Cholesterol')

# Add legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a line
ax.axhline(over_50['chol'].mean(),
                  linestyle='--');
```

# Customization of Plots

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);
# add data
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target']);

#Customize
ax0.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol');
# add legend
ax0.legend(*scatter.legend_elements(),title='Target');
#add mean line
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--')

# add data ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target']);

#Customize ax1
ax1.set(title='Heart Disease and Thalach Levels',
       xlabel='Age',
       ylabel='Thalach');
# add legend
ax1.legend(*scatter.legend_elements(),title='Target');
#add mean line
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--')

#Add title to entire figure
fig.suptitle('Heart Disease and Analysis', fontsize=16,fontweight='bold');
```
